```python
class node:
  def __init__(self, key):
    self.key = key
    self.prev = None
    self.next = None

class linked_list:
  def __init__(self,n):
    self.length = n
    self.head = None
    self.tail = None
    for i in range(1,n+1):
      self.append(i)
    self.tail.next = self.head
    self.head.prev = self.tail

  def append(self,key):
    new_node = node(key)
    if self.head == None:
      self.head = new_node
      self.tail = new_node
    else:
      self.tail.next = new_node
      new_node.prev = self.tail
      self.tail = new_node

  def delete(self,key):
    if self.head == None:
      return
    else:
      curr_node = self.head
      while curr_node.key != key:
        curr_node = curr_node.next
      curr_node.prev.next = curr_node.next
      curr_node.next.prev = curr_node.prev
      self.length -= 1

  def sol(self, k):
    curr_node = self.tail
    print("<", end = "")
    for _ in range(self.length-1):
      for _ in range(k):
        curr_node = curr_node.next
      print(curr_node.key, end = "")
      print(",", end = " ")
      self.delete(curr_node.key)
    print(curr_node.next.key, end = "")
    print(">", end = " ")


n,k = map(int, input().split())
l = linked_list(n)
l.sol(k)
```

```python
def sol(m,n,x,y):
  year = 1
  i,j = 1,1
  while i != x or j != y:
    i = i+1 if i < m else 1
    j = j+1 if j < n else 1
    year += 1
    if i == m and j == n:
      if x==m and y==n:
        return year
      return -1
  return year

k = int(input())
for _ in range(k):
  m,n,x,y = map(int,input().split())
  print(sol(m,n,x,y))

```

```python
from collections import deque

"""
1. 초기에 실내공기와 치즈내부 분리 table[0][0] bfs로
2. 2변 이상 접하는 치즈제거
3. 치즈 내부로 실내공기 침투
4. 2~3 반복
"""
n,m = map(int,input().split())
table = []
for _ in range(n):
  table.append(list(map(int,input().split())))

def bfs(table):
  dy = [0,0,1,-1]
  dx = [1,-1,0,0]
  visit = [[False]*m for _ in range(n)]
  q = deque([(0,0)])
  table[0][0] = 2
  visit[0][0] = True
  while q:
    b,a = q.popleft()
    for i in range(4):
      y,x = b+dy[i],a+dx[i]
      if 0<=y<n and 0<=x<m and not visit[y][x] and table[y][x] != 1:
        table[y][x] = 2
        visit[y][x] = True
        q.append((y,x))

def melt(table,cheese):
  tmp = [arr[:] for arr in table]
  dy = [0,0,1,-1]
  dx = [1,-1,0,0]
  for i in range(n):
    for j in range(m):
      if tmp[i][j] == 1:
        air = 0
        for k in range(4):
          y,x = i+dy[k],j+dx[k]
          if 0<=y<n and 0<=x<m and tmp[y][x] == 2:
            air += 1
        if air >= 2:
          table[i][j] = 2
          cheese -= 1

  return cheese

cheese = 0
for i in range(n):
  for j in range(m):
    if table[i][j] == 1:
      cheese += 1


t = 0
while cheese != 0:
  bfs(table)
  cheese = melt(table,cheese)
  t += 1

print(t)
```
