```python
from collections import deque

n = int(input())
num = list(map(int,input().split()))
m = int(input())

fee = {}

for i in range(n):
  tmp = num.pop()
  if tmp not in num:
    fee[i] = tmp

max_num = 0
stack = deque([(0, m)])
while stack:
  room, budget = stack.popleft()
  for i in range(n):
    if num[i] <= budget:
      stack.append((10*room+i, budget-num[i]))
      max_num = max(10*room+i, max_num)

print(max_num)
```

```python
def prime(n):
  if n <= 1:
    return False

  for i in range(2,round(n**1/2)+1):
    if n%i == 0:
      return False
  return True

t = int(input())
for _ in range(t):
  n = int(input())
  if n <= 2:
    print(2)
  else:
    if n%2 == 0:
      n+=1
    while not prime(n):
      n+=2
    print(n)
```
